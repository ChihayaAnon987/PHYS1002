### 使用方法
1. 修改 `f_0` 为实际测量的共振频率
2. 修改 `V_pp_1` 为「研究振荡频率对电力传输效率的影响」实验中测得的电压峰峰值
3. 修改 `V_pp_2` 为「研究无线电力传输的距离对传输效果影响」实验中测得的电压峰峰值
4. 运行程序，得到两幅曲线图，在程序根目录下会保存为 `amplitude-frequency-curve.png` 和 `voltage-distance-curve.png`

```python
from matplotlib import pyplot as plt

f_0 = 2258.8 # 测得系统实际共振频率，KHz

# 幅度-频率曲线
f = [f_0 - 160, f_0 - 80, f_0 - 50, f_0 - 30, f_0, f_0 + 30, f_0 + 50, f_0 + 80, f_0 + 160] # 频率，KHz
V_pp_1 = [2.32, 4.16, 6.08, 8.72, 10.2, 8.04, 4.92, 2.24, 0.80] # 峰峰值，V

plt.figure(figsize=(10, 6))
plt.plot(f, V_pp_1, 'o-')
plt.xlabel('Frequency (KHz)')
plt.ylabel('$V_pp$ (V)')
plt.title('Amplitude-Frequency Curve')
plt.grid()
plt.savefig('amplitude-frequency-curve.png')
plt.show()

# 电压-距离曲线
distance = [10, 13, 16, 19, 22, 25, 28, 31] # 距离，cm
V_pp_2 = [4.08, 7.68, 11.8, 15.0, 16.6, 15.8, 14.0, 11.0] # 峰峰值，V

plt.figure(figsize=(10, 6))
plt.plot(distance, V_pp_2, 'o-')
plt.xlabel('Distance (cm)')
plt.ylabel('$V_pp$ (V)')
plt.title('Voltage-Distance Curve')
plt.grid()
plt.savefig('voltage-distance-curve.png')
plt.show()
```
