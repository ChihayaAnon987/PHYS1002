```python
import numpy as np
import matplotlib.pyplot as plt
```

# 幅度-频率曲线

```python
frequency = [2.110,2.190,2.220,2.240,2.270,2.300,2.320,2.350,2.430]  # 频率范围, 表二,MHZ
amplitude = [5.40, 7.80, 8.80, 10.2, 14.2, 17.0, 16.0, 11.8, 3.4]  # 振幅,V

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(frequency, amplitude)
plt.scatter(frequency, amplitude, color='red')  # 显示每个点
for i, txt in enumerate(amplitude):
    plt.text(frequency[i], amplitude[i], f'({frequency[i]:.2f}, {amplitude[i]:.2f})', fontsize=8, ha='left', va='bottom')
plt.title('Amplitude-Frequency Curve')
plt.xlabel('Frequency/MHZ')
plt.ylabel('Amplitude/V')
plt.show()
```

# 灯泡电压-距离曲线

```python
distance = [10, 13, 16, 19, 22, 25, 28, 31]  # 距离范围, cm,表四
voltage = [11.2, 15.2, 17.4, 16.4, 13.6, 11.2, 8.40, 6.60]  # 电压,V
plt.figure(figsize=(6, 4))
plt.plot(distance, voltage)
plt.scatter(distance, voltage, color='blue')  # 显示每个点
for i, txt in enumerate(voltage):
    plt.text(distance[i], voltage[i], f'({distance[i]:.2f}, {voltage[i]:.2f})', fontsize=8, ha='left', va='bottom')
plt.title('Voltage-Distance Curve')
plt.xlabel('Distance/cm')
plt.ylabel('Voltage/V')
plt.show()
```
